## Introduction
In this project, we aim to use a thorough data-driven research to determine Pittsburgh's best neighborhood. In order to do this, we have developed a composite metric that takes into account important characteristics from several datasets, such as "arrest.csv," "fire.csv," and "play.csv." Our method combines recreational and safety factors to create a thorough assessment of neighborhood desirability.

The Metrics:


Arrests: A neighborhood is considered safer if there are fewer arrests.

Fire Incidents: A reduced number of fire incidents points to infrastructure maintenance and safety.

Playgrounds: More playgrounds indicate better recreational amenities.

## Arrests 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
file_path = "arrest.csv"
df = pd.read_csv(file_path)

# Count the occurrences of arrests in each neighborhood
neighborhood_arrests = df['INCIDENTNEIGHBORHOOD'].value_counts()

# Select the top 5 neighborhoods with the lowest arrests
top_5_neighborhoods = neighborhood_arrests.tail(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
top_5_neighborhoods.plot(kind='bar', color='skyblue')
plt.title('Top 5 Neighborhoods with Lowest Arrests')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Arrests')
plt.show()
```

## Fire Incidents

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
file_path_fire = "fire.csv"
df_fire = pd.read_csv(file_path_fire)

# Count the occurrences of fire incidents in each neighborhood
neighborhood_fires = df_fire['neighborhood'].value_counts()

# Select the top 5 neighborhoods with the lowest fire incidents
top_5_neighborhoods_fire = neighborhood_fires.tail(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
top_5_neighborhoods_fire.plot(kind='bar', color='orange')
plt.title('Top 5 Neighborhoods with Lowest Fire Incidents')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Fire Incidents')
plt.show()
```

## Playgrounds

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
file_path_play = "play.csv"
df_play = pd.read_csv(file_path_play)

# Count the occurrences of playgrounds in each neighborhood
neighborhood_playgrounds = df_play['neighborhood'].value_counts()

# Select the top 5 neighborhoods with the most playgrounds
top_5_neighborhoods_play = neighborhood_playgrounds.head(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
top_5_neighborhoods_play.plot(kind='bar', color='green')
plt.title('Top 5 Neighborhoods with Most Playgrounds')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Playgrounds')
plt.show()
```
